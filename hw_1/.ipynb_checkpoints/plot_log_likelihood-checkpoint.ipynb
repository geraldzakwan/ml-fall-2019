```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy import special

combination = special.comb(100, 40, exact=True)
print(combination)
# Error
# constant = np.log(combination)
constant = 64.7905624171

theta_values = np.arange(0.0, 1.0, 0.01)
log_likelihood = np.zeros(len(theta_values))
for i in range(0, len(theta_values)):
    b = 40*np.log(theta_values[i]/2+(1/4))
    c = 60*np.log((-1)*(theta_values[i]/2)+(3/4))
    log_likelihood[i] = constant + b + c

# Create the plot
plt.plot(theta_values, log_likelihood)

# Add a title
plt.title('Log likelihood plot')

# Add X and y Label
plt.xlabel('x axis')
plt.ylabel('y axis')

# L(θ|n= 100) = ln(100!60!40!) + 40 ln(12θ+14) + 60 ln(−12θ+34)
plt.savefig('log_likelihood_with_constant.png', bbox_inches='tight')

# Show the plot
plt.show()
```
